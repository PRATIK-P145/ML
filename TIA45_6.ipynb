```python
# Install required library
# !pip install mlxtend

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt

# ---------------------------
# 1. Load the dataset
# ---------------------------
dataset = pd.read_csv('/content/sample_data/Market_Basket_Optimisation.csv', header=None)

# ---------------------------
# 2. Preprocessing: convert dataset to list of transactions
# ---------------------------
transactions = []
for i in range(len(dataset)):
    # drop NaN values for empty cells
    transactions.append([str(item) for item in dataset.iloc[i].dropna()])

# ---------------------------
# 3. Encode transactions
# ---------------------------
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# ---------------------------
# 4. Train Apriori algorithm
# ---------------------------
# min_support=0.01 means items appear in at least 1% of transactions
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)

# ---------------------------
# 5. Generate association rules
# ---------------------------
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.2)
rules = rules.sort_values(by='confidence', ascending=False)

print("Top 10 rules:\n", rules.head(10))

# ---------------------------
# 6. Visualization
# ---------------------------
plt.figure(figsize=(10,6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.6, edgecolor='k')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Association Rules')
plt.show()

# ---------------------------
# 7. Experiment with higher confidence
# ---------------------------
rules_high_conf = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.6)
rules_high_conf = rules_high_conf.sort_values(by='confidence', ascending=False)
print("Rules with high confidence:\n", rules_high_conf.head(10))
```

```python
top_lift_rules = rules.sort_values(by='lift', ascending=False).head(10)
print("\nTop 10 rules by lift:\n")
print(top_lift_rules)

# ---------------------------
# 8. Bar plot for top 10 rules by lift
# ---------------------------
plt.figure(figsize=(12,6))
plt.barh(range(len(top_lift_rules)), top_lift_rules['lift'], color='skyblue')
plt.yticks(range(len(top_lift_rules)),
           [' & '.join([str(x) for x in items]) for items in top_lift_rules['antecedents']])
plt.xlabel('Lift')
plt.title('Top 10 Association Rules by Lift')
plt.gca().invert_yaxis()  # Highest lift on top
plt.show()
```
