```python
df.drop('poisonous',axis=1)
```

```python
X = df.drop(columns = ['poisonous'])
Y = df['poisonous']
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split as tts , cross_val_score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay as cmd , classification_report as cr
```

```python
Xtrain , Xtest , Ytrain , Ytest = tts(X , Y , test_size = 0.25)
```

```python
from sklearn.naive_bayes import GaussianNB as nb
from sklearn.tree import DecisionTreeClassifier as DTC , plot_tree
```

```python
model_gb = nb()
```

```python
model_gb.fit(Xtrain , Ytrain)
```

```python
Ypred_nb = model_gb.predict(Xtest)
```

```python
cmd.from_predictions(Ytest , Ypred_nb)
print(f"classification report : {cr(Ytest , Ypred_nb)}")
```

```python
model_dt = DTC(random_state = 42)
```

```python
model_dt.fit(Xtrain , Ytrain)
```

```python
Ypred_dt = model_dt.predict(Xtest)
```

```python
cmd.from_predictions(Ytest , Ypred_dt)
print(f"Classification report{cr(Ytest , Ypred_dt)}")
```

```python
plt.figure(figsize = (30, 30))
plot_tree(model_dt , filled = True , max_depth = 3 , class_names = X.columns, impurity = False)
plt.show()
```

```python
score_gb = cross_val_score(model_gb , X , Y , cv = 5 , scoring = "accuracy" , n_jobs = -1)
score_dt = cross_val_score(model_dt , X , Y , cv = 5 , scoring = "accuracy" , n_jobs = -1)
```

```python
print(f"Cross val score of gaussianNB : {np.mean(score_gb)}")
print(f"Cross val score of decsionTree : {score_dt}")
```
